## 1. Basic Example of Cluster


Load the data


```python
# importing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from cv2 import kmeans
import seaborn as sns
from sklearn.cluster import KMeans

sns.set()

# Loading the data
data = pd.read_csv(
    "C:/Users/HP 840 G1/Documents/VS Code Projects/Workspace Learning/DataScience/DataScienceInPython/Drawable/6.Country_clusters.csv"
)

print(data)

# plot the data
plt.scatter(data["Longitude"], data["Latitude"])
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

1. Here will will cluster the countries based on their geo position(neumerical data)


```python
# Select the features, ## we are interested in latitude and longitude
x = data.iloc[:, 1:3]
print(x)

# 1. Clustering, where no. of cluster k = 2
kmeans = KMeans(n_clusters=2)
kmeans.fit(x)

# Clustering Results
identified_clusters = kmeans.fit_predict(x)
print(identified_clusters)

# Clustering result adding to the dataframe
data_with_clusters = data.copy()
data_with_clusters["Cluster"] = identified_clusters
data_with_clusters
# it clear that some data are in cluster 0 ane some data are in cluster 1

# plot the data with clusters
plt.scatter(
    data["Longitude"], data["Latitude"], c=data_with_clusters["Cluster"], cmap="rainbow"
)
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()

# 2. Clustering, where no. of cluster k = 3
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)

# Clustering Results
identified_clusters = kmeans.fit_predict(x)
print(identified_clusters)

# Clustering result adding to the dataframe
data_with_clusters = data.copy()
data_with_clusters["Cluster"] = identified_clusters
data_with_clusters
# it clear that some data are in cluster 0 ane some data are in cluster 1

# plot the data with clusters
plt.scatter(
    data["Longitude"], data["Latitude"], c=data_with_clusters["Cluster"], cmap="rainbow"
)
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

2. Here will will cluster the countries based on their language(categorical data)


```python
# Map the data
data_mapped = data.copy()
data_mapped["Language"] = data_mapped["Language"].map(
    {"English": 0, "French": 1, "German": 2}
)
data_mapped

# Select the features
x = data_mapped.iloc[:, 3:4]
print(x)

# 1. Clustering, where no. of cluster k = 2
kmeans = KMeans(n_clusters=2)
kmeans.fit(x)

# Clustering Results
identified_clusters = kmeans.fit_predict(x)
print(identified_clusters)

# Clustering result adding to the dataframe
data_with_clusters = data.copy()
data_with_clusters["Cluster"] = identified_clusters
data_with_clusters
# it clear that some data are in cluster 0 ane some data are in cluster 1

# plot the data with clusters
plt.scatter(
    data["Longitude"], data["Latitude"], c=data_with_clusters["Cluster"], cmap="rainbow"
)
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()

# 2. Clustering, where no. of cluster k = 3
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)

# Clustering Results
identified_clusters = kmeans.fit_predict(x)
print(identified_clusters)

# Clustering result adding to the dataframe
data_with_clusters = data.copy()
data_with_clusters["Cluster"] = identified_clusters
data_with_clusters
# it clear that some data are in cluster 0 ane some data are in cluster 1

# plot the data with clusters
plt.scatter(
    data["Longitude"], data["Latitude"], c=data_with_clusters["Cluster"], cmap="rainbow"
)
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

3. Here will will cluster the countries based on their geo position and language(both neumeric and categorical data)

4. How to choose the number of clusters


Selecting the Number of Cluster

Criterion: draw wcss vs k and find the "Elbow"


```python
# WCSS
kmeans.inertia_  # finds the inertia for a specific kmeans

wcss = []
for i in range(1, 7):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

# The Elbow Method
number_cluster = range(1, 7)
plt.plot(number_cluster, wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("Within-cluster Sum of Square or WCSS")
plt.show()


# The
```

## 2. Market Sagmentation with Cluster Analysis


Market Data Description

 


```python
# Importing necesary Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

sns.set()

# ##Load the data
data = pd.read_csv(
    "C:/Users/HP 840 G1/Documents/VS Code Projects/Workspace Learning/DataScience/DataScienceInPython/Drawable/7.Example.csv"
)

# print(data)

# ## Plot the data
plt.scatter(data["Satisfaction"], data["Loyalty"])
plt.xlabel("Satisfaction")
plt.ylabel("Loyalty")
plt.show()
```

Here from the scatter plot we can have 4 regions




```python
# ##Select the features
x = data.copy()

# ## Clustering
kmeans = KMeans(2)
kmeans.fit(x)

# ##Clustering Result
cluster = x.copy()
cluster["cluster_pred"] = kmeans.fit_predict(x)

# ##Plot the cluster analysis
plt.scatter(
    cluster["Satisfaction"],
    cluster["Loyalty"],
    c=cluster["cluster_pred"],
    cmap="rainbow",
)
plt.xlabel("Satisfaction")
plt.ylabel("Loyalty")
plt.show()
```

Here the result when clustering is based on "Satisfaction"



This problem can be fixed by standadizing the variables




```python
# ##Standadize the vairables
from sklearn import preprocessing

x_scaled = preprocessing.scale(x)

# print(x_scaled)

# ##Take advantage of the Elbow method
wcss = []
for i in range(1, 10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)

# print(wcss)
plt.plot(range(1, 10), wcss)
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

Clustering from the scaled features


```python
cluster_new = x.copy()

plt.figure(figsize=(10, 10))
# ##Taking k=2 to k=6
for i in range(1, 6):
    # ##Clustering
    kmeans = KMeans(i + 1)
    kmeans.fit(x_scaled)
    cluster_new["Cluster_pred"] = kmeans.fit_predict(x_scaled)

    # ##Ploting the clustering
    plt.suptitle("Market Sagmentation")
    plt.subplot(3, 2, i)
    plt.tight_layout(pad=1, w_pad=1, h_pad=1)
    plt.xlabel("Satisfaction")
    plt.ylabel("Loyalty")
    plt.scatter(
        cluster_new["Satisfaction"],
        cluster_new["Loyalty"],
        c=cluster_new["Cluster_pred"],
        cmap="rainbow",
    )
    plt.title("{} Cluster".format(i + 1))
```
