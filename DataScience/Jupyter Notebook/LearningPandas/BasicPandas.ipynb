## Aggregate Statistics (Groupby)


```python
# 'Type 1', 'Type 2', 'Name' ei column gulir prottek ta element er mean Defense er ascending order onujayi
modified_df.groupby(['Type 1', 'Type 2', 'Name']).mean(
).sort_values('Defense', ascending=False)

# notun akta column add korlam
modified_df['count'] = 1

# count column e Type 1 er element er prottekta kotobar ache seta store korlam
modified_df.groupby(['Type 1']).count()['count']
# modified_df.groupby(['Type 1', 'Type 2']).count()['count']
# modified_df
```

## Working with large amounts of data


```python
# notun akta dataframe new_df jar column ager motoi
new_df = pd.DataFrame(columns=df.columns)

# ashol dataframe er theke 5 ta 5 ta kore row niye ak ak ta df er vitore rekhe
# setar Type 1 er element gulir count new_df e concatenate korlam
for df in pd.read_csv('modified_pokemon_data.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()

    new_df = pd.concat([new_df, results])
new_df
```
