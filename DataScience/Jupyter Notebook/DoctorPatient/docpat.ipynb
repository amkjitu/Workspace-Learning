````python
class BanglaTextPreprocessor:
    def __init__(self):
        self.basic_tokenizer = BasicTokenizer()
        
    def normalize_text(self, text):
        """Normalize Bangla text"""
        # Remove extra whitespaces
        text = re.sub(r'\s+', ' ', text.strip())
        
        # Normalize common medical terms
        text = text.replace('ডাঃ', 'ডাক্তার')
        text = text.replace('রোগীঃ', 'রোগী:')
        
        return text
    
    def tokenize(self, text):
        """Tokenize Bangla text"""
        normalized_text = self.normalize_text(text)
        tokens = self.basic_tokenizer.tokenize(normalized_text)
        return tokens
    
    def preprocess_conversation(self, conversation):
        """Preprocess entire conversation"""
        lines = conversation.split('\n')
        processed_lines = []
        
        for line in lines:
            if line.strip():
                normalized = self.normalize_text(line)
                processed_lines.append(normalized)
        
        return '\n'.join(processed_lines)

# Initialize preprocessor
preprocessor = BanglaTextPreprocessor()
print("🔧 Bangla text preprocessor initialized!")
````

````python
# Load Bangla-BERT for NER (we'll simulate NER since medical NER models are specialized)
class BanglaMedicalNER:
    def __init__(self):
        # For demonstration, we'll use rule-based extraction with common medical terms
        # In production, you'd use a fine-tuned model
        
        self.symptoms_keywords = [
            'জ্বর', 'কাশি', 'মাথাব্যথা', 'গলাব্যথা', 'পেটে ব্যথা', 'পেটব্যথা',
            'বমি', 'ডায়রিয়া', 'শ্বাসকষ্ট', 'বুকে ব্যথা', 'হাতে ব্যথা', 'পায়ে ব্যথা'
        ]
        
        self.medications_keywords = [
            'প্যারাসিটামল', 'ওমিপ্রাজল', 'অ্যাজিথ্রোমাইসিন', 'আইবুপ্রোফেন',
            'এমোক্সিসিলিন', 'সিপ্রোফ্লক্সাসিন', 'মেট্রোনিডাজল'
        ]
        
        self.duration_patterns = [
            r'(\d+)\s*(দিন|সপ্তাহ|মাস|ঘন্টা)',
            r'(তিন|চার|পাঁচ|ছয়|সাত|আট|নয়|দশ)\s*(দিন|সপ্তাহ|মাস)',
            r'(এক|দুই|তিন|চার|পাঁচ)\s*(সপ্তাহ|মাস)'
        ]
    
    def extract_entities(self, text):
        """Extract medical entities from Bangla text"""
        entities = {
            'symptoms': [],
            'medications': [],
            'durations': []
        }
        
        # Extract symptoms
        for symptom in self.symptoms_keywords:
            if symptom in text:
                entities['symptoms'].append(symptom)
        
        # Extract medications
        for medication in self.medications_keywords:
            if medication in text:
                entities['medications'].append(medication)
        
        # Extract durations using regex
        for pattern in self.duration_patterns:
            matches = re.findall(pattern, text)
            for match in matches:
                if isinstance(match, tuple):
                    duration = ' '.join(match)
                else:
                    duration = match
                entities['durations'].append(duration)
        
        return entities

# Initialize NER model
ner_model = BanglaMedicalNER()
print("🎯 Bangla Medical NER model initialized!")
````

````python
class BanglaMedicalConversationAnalyzer:
    def __init__(self, preprocessor, ner_model, summarizer=None):
        self.preprocessor = preprocessor
        self.ner_model = ner_model
        self.summarizer = summarizer
    
    def analyze_conversation(self, conversation):
        """Analyze a complete doctor-patient conversation"""
        # Step 1: Preprocess
        processed_text = self.preprocessor.preprocess_conversation(conversation)
        
        # Step 2: Extract entities
        entities = self.ner_model.extract_entities(processed_text)
        
        # Step 3: Generate structured summary
        summary = self.generate_structured_summary(entities, processed_text)
        
        return {
            'original_text': conversation,
            'processed_text': processed_text,
            'entities': entities,
            'summary': summary
        }
    
    def generate_structured_summary(self, entities, text):
        """Generate a structured summary based on extracted entities"""
        summary_parts = []
        
        # Symptoms section
        if entities['symptoms']:
            symptoms_text = ', '.join(entities['symptoms'])
            summary_parts.append(f"লক্ষণ: {symptoms_text}")
        
        # Medications section
        if entities['medications']:
            medications_text = ', '.join(entities['medications'])
            summary_parts.append(f"ওষুধ: {medications_text}")
        
        # Duration section
        if entities['durations']:
            durations_text = ', '.join(entities['durations'])
            summary_parts.append(f"সময়কাল: {durations_text}")
        
        # Generate narrative summary
        if summary_parts:
            structured_summary = ' | '.join(summary_parts)
        else:
            structured_summary = "কোন নির্দিষ্ট তথ্য পাওয়া যায়নি।"
        
        return {
            'structured': structured_summary,
            'entity_count': {
                'symptoms': len(entities['symptoms']),
                'medications': len(entities['medications']),
                'durations': len(entities['durations'])
            }
        }
    
    def batch_analyze(self, conversations):
        """Analyze multiple conversations"""
        results = []
        for i, conv in enumerate(conversations):
            print(f"\n🔍 Analyzing conversation {i+1}...")
            result = self.analyze_conversation(conv)
            results.append(result)
        return results

# Initialize the analyzer
analyzer = BanglaMedicalConversationAnalyzer(preprocessor, ner_model, summarizer)
print("🏥 Bangla Medical Conversation Analyzer ready!")
````

````python
# Utility functions for model evaluation and extension

def evaluate_extraction_accuracy(true_entities, extracted_entities):
    """Calculate precision, recall, and F1 for entity extraction"""
    def calculate_metrics(true_set, extracted_set):
        if not extracted_set:
            return 0, 0, 0
        
        true_positive = len(true_set.intersection(extracted_set))
        precision = true_positive / len(extracted_set) if extracted_set else 0
        recall = true_positive / len(true_set) if true_set else 0
        f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        
        return precision, recall, f1
    
    results = {}
    for entity_type in ['symptoms', 'medications', 'durations']:
        true_set = set(true_entities.get(entity_type, []))
        extracted_set = set(extracted_entities.get(entity_type, []))
        
        precision, recall, f1 = calculate_metrics(true_set, extracted_set)
        results[entity_type] = {
            'precision': precision,
            'recall': recall,
            'f1': f1
        }
    
    return results

def add_custom_medical_terms(ner_model, new_symptoms=None, new_medications=None):
    """Add custom medical terms to the NER model"""
    if new_symptoms:
        ner_model.symptoms_keywords.extend(new_symptoms)
        print(f"✅ Added {len(new_symptoms)} new symptoms")
    
    if new_medications:
        ner_model.medications_keywords.extend(new_medications)
        print(f"✅ Added {len(new_medications)} new medications")
    
    # Remove duplicates
    ner_model.symptoms_keywords = list(set(ner_model.symptoms_keywords))
    ner_model.medications_keywords = list(set(ner_model.medications_keywords))

def export_results_to_csv(results, filename='bangla_medical_analysis.csv'):
    """Export analysis results to CSV"""
    data = []
    
    for i, result in enumerate(results):
        row = {
            'conversation_id': i + 1,
            'symptoms': ', '.join(result['entities']['symptoms']),
            'medications': ', '.join(result['entities']['medications']),
            'durations': ', '.join(result['entities']['durations']),
            'summary': result['summary']['structured'],
            'symptom_count': result['summary']['entity_count']['symptoms'],
            'medication_count': result['summary']['entity_count']['medications'],
            'duration_count': result['summary']['entity_count']['durations']
        }
        data.append(row)
    
    df = pd.DataFrame(data)
    df.to_csv(filename, index=False, encoding='utf-8')
    print(f"📁 Results exported to {filename}")
    return df

print("🛠️ Utility functions created!")
````
